# Web Scraping Pipeline - BaseballReference.com

```python
from bs4 import BeautifulSoup, Comment
import requests 
import pandas as pd
```

```python
tables = []

for year in range(1970, 2022):
    url = f'https://www.baseball-reference.com/leagues/majors/{year}.shtml'
    result = requests.get(url).text
    data = BeautifulSoup(result, 'html.parser')

    comments = data.find_all(string=lambda text: isinstance(text, Comment))

    for each in comments:
        if 'table' in str(each):
            try:
                tables.append(pd.read_html(str(each), attrs={'id':'teams_standard_pitching'})[0])
                break
            except:
                continue
```

```python
for i in range(0, 52):
    tables[i]['YR'] = i
    
tables[0]
```

```python
bb_df = pd.concat(tables)
bb_df
```

```python
bb_df['Tm'].unique
```

```python
bb_filter = bb_df[bb_df['Tm'] != 'League Average']
bb_filter2 = bb_filter[bb_filter['Tm'] != 'Tm']
bb = bb_filter2
```

```python
bb.dropna(subset=['Tm'], inplace=True)
```

```python
bb.to_csv('pitching.csv')
```
